```python
import pandas as pd
import csv
import json
from flask import jsonify
```

```python
#Create connection to database.. if it doesn't exist it should create it

from sqlalchemy import create_engine
engine = create_engine('sqlite:///C:\\Users\\zulim2\\Downloads\\Analytics\\World-Cup-Analysis\\data\\world_cup.db', echo=False)
```

```python
#pull in rankings csv and push it to a new db table

wc_data_df = pd.read_csv("World_Cup_Data_Combined.csv")
wc_data_df.to_sql("rankings_table", con=engine, if_exists='replace', index=False)
wc_data_df
```

```python
#pull in metadata csv and push it to a new db table

wc_metadata_df = pd.read_csv("World_Cup_MetaData.csv")
wc_metadata_df.to_sql("metadata_table", con=engine, if_exists='fail', index=False)
wc_metadata_df
```

```python
#testing the code that the app route "db" will run

data = engine.execute("SELECT * FROM rankings_table").fetchall()
headers = engine.execute("SELECT * FROM rankings_table").keys()

json_data = []

for row in data:
    json_data.append(dict(zip(headers, row)))
```

```python
#outputting the json to a file to work with while building viz

with open('rankings_json.txt', 'w') as outfile:  
    json.dump(json_data, outfile)
```

```python
#outputting the metadata json to a file to work with building viz

metadata_data = engine.execute("SELECT * FROM metadata_table").fetchall()
metadata_headers = engine.execute("SELECT * FROM metadata_table").keys()

metadata_json = []

for row in metadata_data:
    metadata_json.append(dict(zip(metadata_headers, row)))

metadata = json.dumps(metadata_json)

with open('metadata_json.txt', 'w') as outfile:  
    json.dump(metadata_json, outfile)
```
